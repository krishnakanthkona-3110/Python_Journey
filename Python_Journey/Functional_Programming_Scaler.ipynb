1. Function – First Class Citizen

Meaning

In Functional Programming, function is treated like a normal variable

That means function can be:

• stored in variable

• passed as argument

• returned from another function

• stored in data structures

✅ Key Features



# Pure Functions

Same input → Same output

No side effects

# Immutability

Data cannot be modified

New data is created instead

# First-Class Functions

Functions can be passed as arguments

Functions can be returned

# No Shared State

Avoids global variable modification

# Declarative Style

Focus on what to do, not how to do

```python
# Higher Order Function:-
'''
A Higher-Order Function (HOF) in Python is a function that either:
Takes another function as an argument, or
Returns another function as its result.

Built-in Higher-Order Functions in Python:-

map(), filter(), reduce(), sorted()
'''
```

```python
# Develop a function that takes the number and calculates a power of it. x**n
# Just fix 2 as the power, fix 3 as the power
```

```python
def square(x):
    return x * x

def apply_function(func, value):
    return func(value)   # calling the function passed

result = apply_function(square, 5)
print(result)   # Output: 25
```

```python
def outer_function():
    def inner_function():
        return "Hello from inner!"
    return inner_function   # returning a function, not calling it

result = outer_function()
print(result())  # Output: Hello from inner!
```

```python
def power(n):
  def exp(x):
    return x**n
  return exp
```

```python
fix = power(3)
```

```python
fix(4)
```

```python
def su(x,y,*args):
  print('args has:', args)
  res = x+y
  if args:
    res = res + sum(args)
  return res
```

```python
su(1,2,3,4)
```

```python
# elements in args are stored as tuple
```

```python
def create_person(name, age, gender, **extra_info):
  person = {'name': name, 'age': age, 'gender': gender}
  print('kwargs are stored as:', extra_info)
  if extra_info:
    person.update(extra_info)
  return person
```

```python
create_person('suraj','90',gender='Male', hobby='Cricket')
```
