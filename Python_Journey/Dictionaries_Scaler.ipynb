```python
# sort based on the values, without key it would sort based on the key
print(sorted(d.items()))
print(sorted(d.items(),key = lambda x:x[1])) # key is the 2nd column (values)
```

```python
k = 'aaabbcddeff'
result = {}
for i in set(k):
  result[i]= k.count(i)
result
```

```python
#sort based on keys
dict(sorted(result.items()))
```

```python
#sorted based on values
dict(sorted(result.items(), key=lambda x:x[1]))
```

```python
#sorted based on keys
dict(sorted(result.items(), key=lambda x:x[0]))
```

```python
def count_char(string):
  # write your code here
  k={}
  for i in set(string):
      c = string.count(i)
      k[i]=c
  return k
```

```python
count_char('aaabbccccccddhii')
```

```python
k = 'aaabbcddeffff'
chars = []
counts = []

for i in k:
  if i in chars:
    id = chars.index(i)
    counts[id]+=1
  else:
    chars.append(i)
    counts.append(1)
print(chars,counts)
```

```python
#best time complexity - O(n)
k = 'aaabbcddeffff'
res={}
for i in k:
  res[i] = res.get(i,0)+1
res
```
