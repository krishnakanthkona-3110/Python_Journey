## Rest Api: REST API stands for
👉 Representational State Transfer – Application Programming Interface

It’s a way for two systems (or applications) to communicate with each other over the internet using standard web protocols (like HTTP).

First defined in 2000 by computer scientist - Dr. Roy Fielding

# @api_view is a decorator provided by Django REST Framework (DRF).

It is used to turn a normal Django function view into a REST API endpoint that can handle JSON requests and responses properly.

Think of @api_view as a "wrapper" that converts your simple function into a Django REST Framework-powered API endpoint — adding extra features like safety checks, method handling, and response formatting.

image.png

#Mixins:
Mixins are predefined reusable classes provided by DRF that help you implement common CRUD functionalities (Create, Retrieve, Update, Delete) without writing repetitive code.

They work together with generic views (like GenericAPIView) to provide complete view logic in just a few lines.

In short

✅ Mixins = small reusable pieces of CRUD logic
✅ Combine with GenericAPIView to make views modular
✅ Reduce boilerplate code in your API

🔹 Common DRF Mixins

CreateModelMixin	For creating an object (POST)

ListModelMixin	For listing all objects (GET)

RetrieveModelMixin	For retrieving a single object (GET with ID)

UpdateModelMixin	For updating an object (PUT/PATCH)

DestroyModelMixin	For deleting an object (DELETE)

# Authentication:

Note: Don't forget that authentication by itself won't allow or disallow an incoming request, it simply identifies the credentials that the request was made with.

Types of Authentications:
1) Basic Authentication
2) Token Authentication
3) Session Authentication
4) RemoteUser Authentication



# Permissions:
Authentication or identification by itself is not usually sufficient to gain access to information or code. For that, the entity requesting access must have authorization.

The simplest style of permission would be to allow access to any authenticated user, and deny access to any unauthorized user. This corresponds to the **IsAuthenticated** class in Rest framework.
