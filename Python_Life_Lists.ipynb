```python
def print_natural_numbers(n):
    if n == 0:
        return
    print_natural_numbers(n - 1)  # Recursive call first (prints from 1 to n)
    print(n)

# Example usage:
N = 10
print(f"First {N} natural numbers:")
print_natural_numbers(N)
```

```python
l1 = [1,2,3,4,5]
l2 = []
for n in l1:
  if n==2:
    l1.remove(2)
  l2.append(n)
  #print(n)
print(l2)

'''Python’s for loop uses an internal index. After processing index 1, it moves to index 2.
But since you modified the list (l1), index 2 now points to 4 (skipped 3).
So n = 4 → append 4 → l2 = [1, 2, 4].'''
```
